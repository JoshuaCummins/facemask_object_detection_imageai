#Transfer learning for training

```
#Taken from the Documentation for training a custom dataset
# https://imageai.readthedocs.io/en/latest/custom/index.html  
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory='/content/drive/MyDrive/3 class mask detection')
trainer.setTrainConfig(object_names_array=['with_mask', 'without_mask', 'mask_weared_incorrect'], batch_size=4, num_experiments=25, train_from_pretrained_model='/content/drive/MyDrive/3 class mask detection/pretrained-yolov3.h5')
trainer.trainModel()
```

```
#code for using a model for a custom video file.
from imageai.Detection.Custom import CustomVideoObjectDetection
import os

execution_path = os.getcwd()

video_detector = CustomVideoObjectDetection()
video_detector.setModelTypeAsYOLOv3()
video_detector.setModelPath("/content/drive/MyDrive/3 class mask detection/models/detection_model-ex-006--loss-0017.094.h5")
video_detector.setJsonPath('/content/drive/MyDrive/3 class mask detection/json/detection_config.json')
video_detector.loadModel()

video_detector.detectObjectsFromVideo(input_file_path="/content/drive/MyDrive/videos/flip.mp4",
                                        output_file_path=os.path.join(execution_path, "flip"),
                                        frames_per_second=30,
                                        minimum_percentage_probability=40,
                                        log_progress=True)
```
